```python
# Replace 'Not Provided' in EmployeeName to NaN
import numpy as np
df['EmployeeName'].replace('Not provided',np.nan, inplace = True)
df['EmployeeName']
```

```python
# Drop the rows having 5 Missing Values
df.drop(df[df.isnull().sum(axis =1) == 5].index, axis = 1, inplace = True) 
```
