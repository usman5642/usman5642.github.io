```python
# Construct a model that shows the relationship between the "Engine Size", "Horsepower", "Cylinders" and "MPG_city" with the "MSRP" of the cars
df.columns
```

```python
X = df[['EngineSize','Horsepower', 'Cylinders', "MPG_City"]]
y = df['MSRP']
```

```python
# Multiple Linear Regression
lm2 = LinearRegression()
```

```python
lm2.fit(X,y)
```

```python
interc = lm2.intercept_
```

```python
coef = lm2.coef_.round(2)
```

```python
# So the Linear Relationship between ['EngineSize','Horsepower', 'Cylinders', "MPG_City"] and "MSRP" can be expressed with the following Equation
print(f"y = {interc} {coef[0]}*x1 +{coef[1]}*x2+ {coef[2]}*x3 + {coef[3]}*x4")
# OR
print(f"MSRP = {interc} {coef[0]}*EngineSize +{coef[1]}*Horsepower+ {coef[2]}*Cylinders + {coef[3]}*MPG_City")
```

```python
# # Visualizing MLR
# One way to look at the fit of the model is by looking at the distribution plot. 
# We can look at the distribution of the fitted values that result from the model and compare it to the distribution of the actual values.
```

```python
y_hat = lm2.predict(X)
```

```python
plt.figure(figsize = (12,6))
ax1 = sns.distplot(df['MSRP'], hist = False, color = "red", label = "Actual Value")
sns.distplot(y_hat, hist = False, color = 'blue', label = "Predicted Values", ax = ax1)
plt.xlabel("MSRP (in $)")
plt.ylabel("Proportion of Cars")
plt.title("Actual vs Fitted Values for MSRP")
plt.show()
```

```python
# We can see that the fitted values are reasonably close to the actual values since the two distributions overlap. 
# However, there is definitely some room for improvement.
```

```python
# Checking the accuracy of the model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
mse = mean_squared_error(df['MSRP'], y_hat)
```

```python
mse.round(2)
```

```python
r_square = r2_score(y, y_hat)
```

```python
r_square.round(2)
```

```python
# A 0.69 value of R square means that 69% of the variation in the response variable y is explained by the multiple linear regression lm2.
```
