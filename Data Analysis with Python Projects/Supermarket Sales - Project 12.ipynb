```python
# Find The Product Line That Generates The Most Income. 
df.columns
```

```python
df.groupby('Product line')['gross income']. sum().sort_values(ascending = False)
```

```python
# Find The Highest Unit Price In The Product Line. 
df.groupby('Product line')['Unit price'].mean().sort_values(ascending = False)
```

```python
sns.barplot(x = 'Product line', y = 'Unit price', data = df)
plt.xticks(rotation = 75)
plt.show()
```

```python
# Find Different Payment Methods Used By Customers Citywise.
pd.crosstab(df["City"],df["Payment"])
```

```python
df.groupby("City")['Payment'].value_counts()
```

```python
sns.countplot(x = 'City', data = df, hue = 'Payment')
plt.legend(loc = [1.1,1])
```

```python
# Which Product Line Is Purchased In The Highest Quantity?
df.groupby('Product line')['Quantity'].sum().plot(kind = 'bar')
```

```python
diction = {0:'Mon',1:'Tue',2:'Wed', 3:'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
```

```python
# Display Daily Sales by Day of the Week
df['day_of_week'] = df['Date'].dt.dayofweek.map(diction)
```

```python
df['day_of_week'].value_counts().plot(kind = 'bar')
```

```python
# What Will Be the Highest Months for Sales
month_mapping = {1:'Jan', 2:'Feb', 3:'Mar'}
```

```python
df['month'] = df['Date'].dt.month.map(month_mapping)
```

```python
df['month'].value_counts().plot(kind = 'bar')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
