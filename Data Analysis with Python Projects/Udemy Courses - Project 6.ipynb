```python
import seaborn as sns
sns.heatmap(df.isnull())
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
# Find Out Number of Courses Per Subjects
df.columns
```

```python
df['subject'].value_counts()
```

```python
import matplotlib.pyplot as plt
sns.countplot(x = df['subject'])
plt.xlabel("Subjects", fontsize = 13)
plt.ylabel("Num of Courses per Subject", fontsize = 13)
plt.xticks(rotation = 60)
plt.show()
```

```python
# For Which Levels, Udemy Courses Providing The Courses
df['level'].value_counts()
```

```python
sns.countplot(x = df['level'])
plt.xlabel("Course Levels", fontsize = 13)
plt.ylabel("Num of Courses per Level", fontsize = 13)
plt.xticks(rotation = 65)
plt.show()
```

```python
# Display The Count of Paid and Free Courses 
df['is_paid'].value_counts()
```

```python
sns.countplot(x= 'is_paid', data =df)
plt.xlabel("Paid/Unpaid", fontsize = 13)
plt.ylabel("Number of Courses", fontsize = 13)
plt.xticks(rotation =60)
plt.show()
```

```python
# Which Course Has More Lectures (Free or Paid)?
df.groupby('is_paid')['num_lectures'].sum()
```

```python
# Which Courses Have A Higher Number of Subscribers Free or Paid?
df.groupby('is_paid')['num_subscribers'].mean()
```

```python
sns.barplot(x = 'is_paid', y = 'num_subscribers', data = df)
```

```python
# Which Level Has The Highest Number of Subscribers?
df.groupby('level')['num_subscribers'].max().sort_values(ascending=False).head(1)
```

```python
sns.barplot(x = 'level', y = 'num_subscribers', data = df)
plt.xlabel("Level", fontsize = 13)
plt.ylabel("Number of Subscribers", fontsize = 13)
plt.xticks(rotation =60)
plt.show()
```

```python
# Find the Most Popular Course Title
df.groupby('course_title')['num_subscribers'].max().sort_values(ascending = False).head(1)
```

```python
 # Display the 10 Most Popular Courses As Per Number of Subscribers
val = df.sort_values(by = 'num_subscribers', ascending = False).head(10)
```

```python
sns.barplot(x = 'num_subscribers', y ='course_title', data = val)
```

```python
# Find The Course Having The Highest Number of Reviews
df.groupby('course_title')['num_reviews'].max().head(1)
```

```python
df2 = df.sort_values(by = 'num_reviews', ascending = False).head(10)
```

```python
sns.barplot(y = 'course_title', x = 'num_reviews', data =df2)
```

```python
# Does Price Affect the Number of Reviews? Both are numerical variables so we can use scatterplot
plt.figure(figsize = (15,5))
sns.scatterplot(y = 'num_reviews', x = 'price', data = df)
```

```python
# Find Total Number of Courses Related To Python
len(df[df['course_title'].str.contains("Python", case = False)])
```

```python
# Display 10 Most Popular Python Courses As Per Number of Subscribers
df3 = df[df['course_title'].str.contains("python", case = False)].sort_values(by = 'num_subscribers', ascending = False).head(10)
```

```python
sns.barplot(y = 'course_title', x = 'num_subscribers', data = df3)
```
