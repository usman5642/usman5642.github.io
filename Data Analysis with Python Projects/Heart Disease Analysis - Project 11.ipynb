```python
# How Many People Have Heart Disease, And How Many Don't Have Heart Disease In This Dataset?
df['target'].value_counts()
# 1 showing the presence of Heart Disease and 0 showing its absence
```

```python
sns.countplot(x = 'target', data = df)
plt.xticks([0,1], ["No Heart Disease", "Heart Disease"])
```

```python
# Find Count of  Male & Female in this Dataset
df['sex'].value_counts()
# 1 showing Male 0 showing Female
```

```python
sns.countplot(x = 'sex', data = df)
plt.xticks([0,1],["Female","Male"])
plt.show()
```

```python
# Find Gender Distribution According to The Target Variable
sns.countplot(x = 'sex', hue = 'target', data = df)
plt.xticks([0,1], ["Female", "Male"])
plt.legend(labels = ["Non-Disease","Disease"])
plt.show()
```

```python
# Check Chest Pain Type
df.columns
```

```python
df['cp'].value_counts()
```

```python
sns.countplot(x = 'cp', data = df)
plt.xticks([0,1,2,3],["Typical Angina", "Atypical Angina", "Non - Anginal Pain", "Asymptomatic"],rotation = 45)
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.show()
```

```python
# Show The Chest Pain Distribution As Per Target Variable
sns.countplot(x = 'cp', data = df, hue ='target')
plt.xticks([0,1,2,3],["Typical Angina", "Atypical Angina", "Non - Anginal Pain", "Asymptomatic"],rotation = 45)
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.legend(labels = ["No-Disease", "Disease"])
plt.show()
```

```python
# Show Fasting Blood Sugar Distribution According To Target Variable
df.columns
```

```python
sns.countplot(x = 'fbs', data = df, hue ='target')
plt.xlabel("Fasting Blood Sugar Level")
plt.ylabel("Count")
plt.legend(labels = ["No-Disease", "Disease"])
plt.show()
```

```python
 # Check Resting Blood Pressure Distribution
plt.hist(x = 'trestbps',data = df)
plt.xlabel("Resting Blood Pressure")
plt.ylabel("Count")
plt.show()
```

```python
# Compare Resting Blood Pressure As Per Sex Column
g = sns.FacetGrid(data = df, hue = 'sex', aspect =4)
g.map(sns.kdeplot,'trestbps', fill = True)
plt.legend(labels = ['Male', 'Female'])
plt.xlabel("Resting Blood Pressure")
plt.show()
```
