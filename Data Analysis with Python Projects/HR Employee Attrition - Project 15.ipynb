```python
# Classify employees based on their total working years into categories like "Junior," "Mid-level," and "Senior."
(df['TotalWorkingYears'].max() - df['TotalWorkingYears'].min())/3
# So we will create groups with 13 years of gap
```

```python
df['TotalWorkingYears'].min()
```

```python
df['TotalWorkingYears'].dtypes
```

```python
def job_level(x):
    if (x>=0) & (x<=13):
        return "Junior"
    elif (x>13) & (x<27):
        return "Mid-Level"
    else:
        return "Senior"
```

```python
df['Level'] = df['TotalWorkingYears'].apply(job_level)
df['Level'].value_counts()
```

```python
# Examine the Attriton by the Level of the Employee (i.e Junior, Mid-Level or Senior)
df.groupby('Level')['Attrition'].value_counts()
```

```python
# Creating regression models to check which model is the best to predict an employee's monthly income.
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
X = df[['TotalWorkingYears']]
y = df['MonthlyIncome']
```

```python
lm.fit(X,y)
```

```python
intercept = lm.intercept_.round(2)
```

```python
coef =lm.coef_[0].round(2)
```

```python
y_hat = lm.predict(X)
```

```python
# So, total working years and monthly income are related by the equation:
print(f"Monthly Income = {intercept} + {coef}*Total Working Years")
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
r_square1 = r2_score(y, y_hat).round(3)
```

```python
mean_error1 = mean_squared_error(y,y_hat).round(2)
```

```python
# An r-square value of 0.597 shows that 59.7% of the variation in y is explained by this single linear regression model.
# But the spread of residuals across the x-axis is not random and suggests a linear model would not be a good fit
```

```python
import numpy as np
```

```python
x = df['TotalWorkingYears']
```

```python
# Carrying out polynomial regression
f = np.polyfit(x,y,3)
p = np.poly1d(f)
```

```python
print(p)
```

```python
r_square2 = r2_score(y,p(x)).round(3)
```

```python
r_square2
```

```python
mean_error2 = mean_squared_error(y,p(x)).round(2)
```

```python
mean_error2
```
